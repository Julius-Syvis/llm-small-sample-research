```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
def plot_base(model_names, models, col_name, zero_val=None):
    plt.figure(figsize=(12, 6))
    num_idxs = []
    
    for i, model_name in enumerate(model_names):
        idxs = [i for (i, m) in enumerate(models) if model_name in m.model.values]
        num_idxs.append(len(idxs))
        vals = [models[j][models[j].model == model_name][col_name].values[0] for j in idxs]
        
        if zero_val is not None:
            vals[0] = zero_val
            
        x = np.array(data_sizes)[idxs]
        
        model_is_large = 'large' in model_name
        model_is_canine = 'canine' in model_name
        
        if model_is_large:
            linestyle = 'dashdot'
        elif model_is_canine:
            linestyle = ':'
        else:
            linestyle = 'solid'
            
        linewidth = 1
        
        plt.plot(x, vals, label=model_name, linewidth=linewidth, linestyle=linestyle)
        plt.xscale('symlog')
    
    plt.xlabel("num samples")
    plt.ylabel(col_name)
    plt.legend()
    plt.show()
    
    print(len(num_idxs), num_idxs)
```

```python
plot_base(model_names, ags, 'accuracy', 0.25)
```
